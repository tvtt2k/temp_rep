```python


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 0.  put both classifiers in **one** tidy DataFrame  (if you havenâ€™t yet)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df1['classifier'] = 'vanilla'      # <- give each its label
df2['classifier'] = 'permuted'

df = pd.concat([df1, df2], ignore_index=True)
# df columns: run, epoch, train_acc, test_acc, classifier

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 1.  aggregate:  mean  +  standard-error (SE = SD / âˆšn) at each epoch
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
summary = (
    df
    .groupby(['classifier', 'epoch'])
    .agg(train_mean=('train_acc', 'mean'),
         test_mean =('test_acc',  'mean'),
         train_se  =('train_acc', lambda x: x.std(ddof=1) / np.sqrt(len(x))),
         test_se   =('test_acc',  lambda x: x.std(ddof=1) / np.sqrt(len(x))))
    .reset_index()
)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# 2.  plot the two classifiers on the same figure with error bars
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig, ax = plt.subplots(figsize=(8, 5))

styles = {
    'vanilla'  : dict(ls='-',  marker='o'),
    'permuted' : dict(ls='--', marker='s')
}

for cls, sty in styles.items():
    sub = summary[summary['classifier'] == cls]

    # train curve Â± SE
    ax.errorbar(sub['epoch'], sub['train_mean'],
                yerr=sub['train_se'],
                fmt=sty['ls'] + sty['marker'],
                capsize=4, label=f'{cls.capitalize()} â€“ train')

    # test curve Â± SE   (make marker hollow so the pair is distinguishable)
    ax.errorbar(sub['epoch'], sub['test_mean'],
                yerr=sub['test_se'],
                fmt=sty['ls'] + sty['marker'],
                mfc='none', capsize=4, label=f'{cls.capitalize()} â€“ test')

ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.set_title('Perceptron vs Permuted Perceptron â€“ mean Â± SE over runs')
ax.grid(True, linestyle='--', linewidth=0.4)
ax.legend()
plt.tight_layout()
plt.show()
```

Why might this be the case?

If a perceptron sees all +1s first and all â€“1s afterwards, it quickly locks onto the initial class, makes almost no corrective updates until the class switch, then over-corrects and repeats the mistake in reverse. Net result: only a handful of informative updates per epoch.
By permuting we can avoid this order-induced bias.
